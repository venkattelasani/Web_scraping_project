# Cleaning the Data.

```python
df.isnull().sum()
```

##### The columns  Discount%, and Selling_Price contain the same missing values across the same rows. Since these columns represent related attributes of a product, where the Selling_Price is typically derived from the Original_Price by applying the Discount%, it would not be meaningful to impute the missing values with statistical methods like the mean or median. Replacing the missing values would distort the relationship between these columns and lead to inaccurate results. 

```python
df = df[~df[[ 'Discount%', 'Selling_Price']].isnull().all(axis=1)]
```

```python
df
```

```python
df.info()
```

```python
df.info()
```

#### changing the datatype for perticular columns.

```python
df.loc[:,'Battery_Life_in_HRS'] = pd.to_numeric(df['Battery_Life_in_HRS'], errors='coerce')
df.loc[:, "Selling_Price"] = df["Selling_Price"].astype(str).str.replace(",", "").astype(float)
```

#### filling null values by grouping with brand mean.

```python
import numpy as np

def fill_with_mean(group):
    filled = group.fillna(group.mean() if not group.dropna().empty else np.nan)
    return filled.round(2)  

columns_to_fill = ['Battery_Life_in_HRS', 'Rating']

for col in columns_to_fill:
    df[col] = df.groupby('Brand')[col].transform(fill_with_mean)
```

#### If Microphone_ENC is present then replacing with FeatureAvaliable else not.

```python
df.loc[:, 'Microphone_ENC'] = df['Microphone_ENC'].notna().map({True: 'FeatureAvailable', False: 'FeatureNotAvailable'}).astype('category')
```

```python
df.loc[:, 'Battery_Life_in_HRS'] = df['Battery_Life_in_HRS'].fillna(df['Battery_Life_in_HRS'].mean())
```

# Data Analysis and Visualization (EDA) 

### Uni-Variate Analysis on continuous data.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

continuous_vars = ["Battery_Life_in_HRS", "Selling_Price", "Discount%", "Rating"] 


for col in continuous_vars:
    print(f"\n{col}:")
    print(f"Mean: {df[col].mean():.2f}, Median: {df[col].median():.2f}, Mode: {df[col].mode()[0]:.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}, Variance: {df[col].var():.2f}")
    
    plt.figure(figsize=(16, 6))
    
    # Box Plot
    plt.subplot(1, 3, 1)
    sns.boxplot(x=df[col], color="lightblue")
    plt.title(f"Boxplot of {col}")
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    # Histogram with KDE
    plt.subplot(1, 3, 2)
    sns.histplot(df[col], kde=True, bins=30, color="blue")
    plt.title(f"Histogram of {col}")
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    # Violin Plot
    plt.subplot(1, 3, 3)
    sns.violinplot(x=df[col], color="lightgreen")
    plt.title(f"Violin Plot of {col}")
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()


```

# Categorical vs Categorical Analysis (Crosstab & Countplot)

```python
def plot_crosstab(df, col1, col2):
    ctab = pd.crosstab(df[col1], df[col2])
    plt.figure(figsize=(8, 6))
    sns.heatmap(ctab, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Crosstab of {col1} vs {col2}")
    plt.xlabel(col2)
    plt.ylabel(col1)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.show()

for col in ["Product_Name", "Drivers", "Microphone_ENC", "Additional_Features"]:
    plot_crosstab(df, col, "Brand")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_stacked_bar_counts(col1, col2, df):
    ctab = pd.crosstab(df[col1], df[col2])  
    ax = ctab.plot(kind="bar", stacked=True, colormap="viridis", figsize=(10, 6), edgecolor="black")

    plt.title(f"{col1} Distribution by {col2}", fontsize=14, fontweight="bold")
    plt.xlabel(col1, fontsize=12)
    plt.ylabel("Count", fontsize=12)  

    # X-axis and Y-axis formatting
    plt.xticks(rotation=45, ha="right", fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Legend formatting
    plt.legend(title=col2, bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=10)

    plt.show()

# Example usage:
plot_stacked_bar_counts("Rating", "Selling_Price", df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_stacked_bar_counts(col1, col2, df):
    ctab = pd.crosstab(df[col1], df[col2])  # Value counts instead of percentages
    ax = ctab.plot(kind="bar", stacked=True, colormap="viridis", figsize=(10, 6), edgecolor="black")

    # Title and labels
    plt.title(f"{col1} Distribution by {col2}", fontsize=14, fontweight="bold")
    plt.xlabel(col1, fontsize=12)
    plt.ylabel("Count", fontsize=12)  # Changed from "Percentage (%)" to "Count"

    # X-axis and Y-axis formatting
    plt.xticks(rotation=45, ha="right", fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Legend formatting
    plt.legend(title=col2, bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=10)

    plt.show()

# Example usage:
plot_stacked_bar_counts("Rating", "Selling_Price", df)
```

```python

def plot_stacked_bar(col1, col2, df):
    ctab = pd.crosstab(df[col1], df[col2], normalize='index') * 100  # Convert to percentage
    ax = ctab.plot(kind="bar", stacked=True, colormap="viridis", figsize=(10, 6), edgecolor="black")

    # Title and labels
    plt.title(f"{col1} Distribution by {col2}", fontsize=14, fontweight='bold')
    plt.xlabel(col2, fontsize=12)
    plt.ylabel("Percentage (%)", fontsize=12)

    # X-axis and Y-axis formatting
    plt.xticks(rotation=45, ha="right", fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis="y", linestyle="--", alpha=0.7)

    # Legend formatting
    plt.legend(title=col1, bbox_to_anchor=(1.05, 1), loc="upper left", fontsize=10)

    plt.show()

# Example usage:
plot_stacked_bar("Rating", "Selling_Price", df)
plot_stacked_bar("Microphone_ENC", "Brand", df)
```
